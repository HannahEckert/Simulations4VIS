```python
import pandas as pd
import numpy as np
```

```python
iteration_1_ndcg_per_user = pd.read_csv("experiments/babyLFM2b1k/output/iteration_1_ndcg_per_user.tsv", sep="\t")
demographics = pd.read_csv("experiments/babyLFM2b1k/input/demographics_raw.tsv", sep="\t")
interactions= pd.read_csv("experiments/babyLFM2b1k/input/dataset.inter", sep="\t")
```

```python
#change colname "_id to "user_id" in demographic in ndcg
demographics.rename(columns={"_id": "user_id"}, inplace=True)
```

```python
#merge the two dataframes on user_id
df = pd.merge(iteration_1_ndcg_per_user, demographics, on="user_id")
```

```python
df["US"] = df["country"] == "US"
```

```python
df["old"] = df["age_on_2013_10_31"] > 25
```

```python
df.head()
```

```python
df.groupby("gender")["ndcg@10"].mean()
```

```python
df.groupby("gender")["ndcg@10"].count()
```

```python
df.groupby("US")["ndcg@10"].mean()
```

```python
df.groupby("old")["ndcg@10"].mean()
```

```python
interactions.rename(columns={"user_id:token": "user_id", "item_id:token": "item_id"}, inplace=True)
```

```python
activity = interactions.groupby("user_id")["item_id"].count()
```

```python
df = df.merge(activity, on="user_id")
```

```python
df.rename(columns={"item_id": "activity"}, inplace=True)
```

```python
df["active"] = df["activity"] > 25
```

```python
df.groupby("active")["ndcg@10"].mean()
```
